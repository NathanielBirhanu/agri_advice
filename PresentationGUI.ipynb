```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Load your dataset
url = "Price_Agriculture_commodities_Week.csv"
df = pd.read_csv(url)

# Initialize the Tkinter application
root = tk.Tk()
root.title("Data Visualization and Recommendation GUI")

# Create a frame for the plot
frame = tk.Frame(root)
frame.pack(fill=tk.BOTH, expand=1)

# Function to clear the frame
def clear_frame():
    for widget in frame.winfo_children():
        widget.destroy()

# Function to plot price distributions
def plot_price_distributions():
    clear_frame()
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    sns.histplot(df['Min Price'], kde=True, ax=axes[0])
    axes[0].set_title('Distribution of Min Price')

    sns.histplot(df['Max Price'], kde=True, ax=axes[1])
    axes[1].set_title('Distribution of Max Price')

    sns.histplot(df['Modal Price'], kde=True, ax=axes[2])
    axes[2].set_title('Distribution of Modal Price')

    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=1)

# Function to plot category distributions
def plot_category_distributions():
    clear_frame()
    fig, axes = plt.subplots(3, 2, figsize=(18, 12))

    categorical_columns = ['State', 'District', 'Market', 'Commodity', 'Variety', 'Grade']
    for i, var in enumerate(categorical_columns, 1):
        ax = axes[(i-1)//2, (i-1)%2]
        top_categories = df[var].value_counts().nlargest(10).index
        sns.countplot(y=df[df[var].isin(top_categories)][var], order=top_categories, ax=ax)
        ax.set_title(f'Top 10 of {var}')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

    plt.tight_layout()
    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=1)

# Function to display data for a specific district
def display_district_data():
    district = district_entry.get().strip()
    if not district:
        messagebox.showerror("Input Error", "Please enter a district name.")
        return
    if district not in df['District'].unique():
        messagebox.showerror("Error", f"District '{district}' not found in dataset.")
        return

    clear_frame()
    district_data = df[df['District'] == district]
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    sns.histplot(district_data['Min Price'], kde=True, ax=axes[0])
    axes[0].set_title(f'Min Price in {district}')

    sns.histplot(district_data['Max Price'], kde=True, ax=axes[1])
    axes[1].set_title(f'Max Price in {district}')

    sns.histplot(district_data['Modal Price'], kde=True, ax=axes[2])
    axes[2].set_title(f'Modal Price in {district}')

    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=1)

# Function to display data for a specific commodity
def display_commodity_data():
    commodity = commodity_entry.get().strip()
    if not commodity:
        messagebox.showerror("Input Error", "Please enter a commodity name.")
        return
    if commodity not in df['Commodity'].unique():
        messagebox.showerror("Error", f"Commodity '{commodity}' not found in dataset.")
        return

    clear_frame()
    commodity_data = df[df['Commodity'] == commodity]
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    sns.histplot(commodity_data['Min Price'], kde=True, ax=axes[0])
    axes[0].set_title(f'Min Price for {commodity}')

    sns.histplot(commodity_data['Max Price'], kde=True, ax=axes[1])
    axes[1].set_title(f'Max Price for {commodity}')

    sns.histplot(commodity_data['Modal Price'], kde=True, ax=axes[2])
    axes[2].set_title(f'Modal Price for {commodity}')

    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=1)

    # Display recommendations
    recommend_similar_commodities(commodity)

# Function to recommend similar commodities
def recommend_similar_commodities(commodity):
    commodity_data = df[df['Commodity'] == commodity]
    avg_modal_price = commodity_data['Modal Price'].mean()

    df['Avg Modal Price'] = df.groupby('Commodity')['Modal Price'].transform('mean')
    similar_commodities = df[df['Commodity'] != commodity].drop_duplicates('Commodity')
    similar_commodities['Price Difference'] = abs(similar_commodities['Avg Modal Price'] - avg_modal_price)
    recommended_commodities = similar_commodities.nsmallest(5, 'Price Difference')['Commodity']

    recommendation_label = tk.Label(frame, text=f"Recommended Commodities Similar to {commodity}:", font=("Arial", 14))
    recommendation_label.pack(pady=10)

    for rec_commodity in recommended_commodities:
        rec_label = tk.Label(frame, text=rec_commodity, font=("Arial", 12))
        rec_label.pack()

# Create input fields and buttons for user input
input_frame = tk.Frame(root)
input_frame.pack(fill=tk.X, side=tk.TOP, padx=5, pady=5)

# District input
district_label = tk.Label(input_frame, text="Enter District:")
district_label.pack(side=tk.LEFT, padx=5)

district_entry = tk.Entry(input_frame)
district_entry.pack(side=tk.LEFT, padx=5)

district_button = ttk.Button(input_frame, text="Display District Data", command=display_district_data)
district_button.pack(side=tk.LEFT, padx=5)

# Commodity input
commodity_label = tk.Label(input_frame, text="Enter Commodity:")
commodity_label.pack(side=tk.LEFT, padx=5)

commodity_entry = tk.Entry(input_frame)
commodity_entry.pack(side=tk.LEFT, padx=5)

commodity_button = ttk.Button(input_frame, text="Display Commodity Data", command=display_commodity_data)
commodity_button.pack(side=tk.LEFT, padx=5)

# Create buttons to trigger the different plots
button_frame = tk.Frame(root)
button_frame.pack(fill=tk.X, side=tk.TOP, padx=5, pady=5)

price_button = ttk.Button(button_frame, text="Plot Price Distributions", command=plot_price_distributions)
price_button.pack(side=tk.LEFT, padx=5, pady=5)

category_button = ttk.Button(button_frame, text="Plot Category Distributions", command=plot_category_distributions)
category_button.pack(side=tk.LEFT, padx=5, pady=5)

# Run the Tkinter event loop
root.mainloop()
```
